```python
import threading
import requests
from bs4 import BeautifulSoup
url=['https://python.langchain.com/v0.2/docs/introduction/',
'https://python.langchain.com/v0.2/docs/concepts/',
'https://python.langchain.com/v0.2/docs/tutorials/']

def fetch_content(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        text = soup.get_text()
        print(f"Content from {url}:\n{text[:500]}...\n")  # Print first 500 characters
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")

threads = []
for link in url:
    thread = threading.Thread(target=fetch_content, args=(link,))
    threads.append(thread)
    thread.start()

for thread in threads:
    thread.join()

```

```python
'''Now we write a code implementing factorial of a number using multiprocessing'''
import multiprocessing
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1) 
if __name__ == '__main__':
    number = 5
    with multiprocessing.Pool() as pool:
        result = pool.apply(factorial, args=(number,))
        print(f"Factorial of {number} is {result}")



```
